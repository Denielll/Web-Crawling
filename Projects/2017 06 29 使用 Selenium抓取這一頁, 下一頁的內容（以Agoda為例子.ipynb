```python
# Agoda進去後有個類似廣告的彈跳視窗  先點掉, 程式才會執行下一頁
# 此行是用 Selenium IDE 做出來後 觀察到的
browser.find_element_by_link_text(u"晚點再選日期 >").click()
```

```python
g = []
for i in range(10):
    try:
        soup = BeautifulSoup(browser.page_source)
        for i in soup.find_all("h3","hotel-name"):
            g.append(i.text)  #注意, 這裡要用 text, 不要用string
        time.sleep(1)
        browser.find_element_by_id("paginationNext").click() #點擊下一頁。這行要靠Selenium IDE完成
    except:
        pass
```
